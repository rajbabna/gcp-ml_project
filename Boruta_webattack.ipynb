```python
import pandas as pd
from seaborn import swarmplot
import matplotlib.pyplot as plt

# Load your dataset from the CSV file
df = pd.read_csv('Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv')

# Print the columns to see what's available
print("Columns in the DataFrame:", df.columns)

# Specify the columns you want to include in the swarmplot
columns_to_plot = [' Destination Port', ' Flow Duration', ' Label']

# Remove leading spaces from column names
columns_to_plot = [column.strip() for column in columns_to_plot]

# Melt the DataFrame
data_to_plot = pd.melt(df[columns_to_plot],
                       id_vars=' Label',
                       var_name='features',
                       value_name='value')

# Create a swarmplot
plt.figure(figsize=(12, 8))
swarmplot(data=data_to_plot, x='features', y='value', hue=' Label')
plt.title('Swarmplot for Selected Columns')
plt.show()
```
